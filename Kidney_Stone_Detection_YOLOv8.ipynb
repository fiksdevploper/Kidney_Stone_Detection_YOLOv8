# Analisa Data

```python
import pandas as pd
import numpy as np
import PIL
from PIL import Image
from IPython.display import display
import matplotlib.pyplot as plt
from glob import glob
import random
import cv2
import warnings
warnings.simplefilter('ignore')
```

```python
root_path = '/content/drive/MyDrive/Kidney Stone Images with Bounding Box Annotations datasests/test/images/*'
num_samples = 4
images_data = glob(root_path)
random_image = random.sample(images_data, num_samples)

plt.figure(figsize=(12,10))
for i in range(num_samples):
    plt.subplot(2,2,i+1)
    plt.imshow(cv2.imread(random_image[i]))
    plt.axis('off')
```

```python
images = []
for i in range(num_samples):
    yolo_outputs = model.predict(random_image[i])
    output = yolo_outputs[0]
    box = output.boxes
    names = output.names
    print('**********************')
    for j in range(len(box)):
        labels = names[box.cls[j].item()]
        coordinates = box.xyxy[j].tolist()
        confidence = np.round(box.conf[j].item(), 2)
        print(f'In this image {len(box)} stones has been detected.')
        print(f'Stone {j + 1} is: {labels}')
        print(f'Coordinates are: {coordinates}')
        print(f'Confidence is: {confidence}')
        print('-------')

    # Store the image in the 'images' list
    images.append(output.plot()[:, :, ::-1])
```

```python
plt.figure(figsize=(12,10))
for i, img in enumerate(images):
    plt.subplot(2, 2, i + 1)
    plt.imshow(img)
    plt.axis('off')
```

# Test Prediksi Ai

```python
from google.colab import files
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
import os

# 1. Panggil fungsi upload
print("Silakan pilih gambar dari komputer Anda:")
uploaded = files.upload()

# 2. Ambil nama file yang baru saja diupload
file_name = list(uploaded.keys())[0]

# 3. Muat model
model = YOLO('/content/runs/detect/train-2/weights/best.pt')

# 4. Jalankan prediksi
results = model.predict(source=file_name, conf=0.5, save=True)

# 5. Cari hasil deteksi (YOLO biasanya menaruh hasil di folder runs/detect/predict/)
# Dynamically get the save directory from the results object
result_save_dir = results[0].save_dir
result_img_path = os.path.join(result_save_dir, file_name)

# 6. Tampilkan hasil
result_img = cv2.imread(result_img_path)

if result_img is not None:
    plt.figure(figsize=(10, 10))
    plt.imshow(cv2.cvtColor(result_img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
else:
    print(f"Error: Could not load the image from {result_img_path}. Please check the path and file integrity.")
```
